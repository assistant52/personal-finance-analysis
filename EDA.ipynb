Obj: Answer the following questions: 
- What do you spend the most money on?
- Compare your income and expenses
- Which account do you use most often, and for what purpose?

```python
import pandas as pd

df = pd.read_csv('C:\\Users\\Admin\\personal-finance-analysis\\data\\personal_transactions.csv')

# Очистка
df['Date'] = pd.to_datetime(df['Date'])
df['Month']   = df['Date'].dt.to_period('M')
df['Year']    = df['Date'].dt.year
df['Weekday'] = df['Date'].dt.day_name()

df_clean = df[df['Category'] != 'Credit Card Payment'].copy()

expenses = df_clean[df_clean['Transaction Type'] == 'debit'].copy()
income   = df_clean[df_clean['Transaction Type'] == 'credit'].copy()
```

Q1: What do you spend your money on?

```python
by_category = (expenses.groupby('Category')['Amount'].sum().sort_values(ascending=False))
print(by_category)
```

```python
avg_by_category = (expenses.groupby('Category')['Amount'].mean().sort_values(ascending=False))
print(avg_by_category)
```

```python
avg_by_weekday = (expenses.groupby('Weekday')['Amount'].mean().sort_values(ascending=False))
print(avg_by_weekday)
```

So, to answer the first question, we can say that $24,754.50 was spent on Mortgage & Rent, this category was also the most expensive on average, and the most money was spent on average on Friday.

Q2: Compare income and expenses

```python
monthly_expenses = expenses.groupby('Month')['Amount'].sum()
monthly_income = income.groupby('Month')['Amount'].sum()

balance = monthly_income - monthly_expenses

print("Monthly expenses:")
print(monthly_expenses)
```

```python
print("Monthly income:")
print(monthly_income)
```

```python
print("Monthly balance:")
print(balance)
```

```python
negative_months = balance[balance < 0]
positive_months = balance[balance > 0]

print("Months with negative balance:")
print(negative_months)
print("Months with positive balance:")
print(positive_months)
print(f"Worst month: {balance.idxmin()} with balance {balance.min():.2f}")
print(f"Best month: {balance.idxmax()} with balance {balance.max():.2f}")
```

Over 21 months of observation, 19 months were profitable, and only two were in the red: May 2018 (-$6,507) and June 2019 (-$6,912). This indicates that the budget is generally managed stably, although these two months are outliers.

Q3:Which account do you use most often, and for what purpose?

```python
by_account_count = (expenses.groupby('Account Name')['Amount'].count().sort_values(ascending=False))

print("Number of transactions :")
print(by_account_count)
```

```python
by_account_sum = (expenses.groupby('Account Name')['Amount'].sum().sort_values(ascending=False))

print("Total amount spent :")
print(by_account_sum)
```

```python
by_account_category = (expenses.groupby(['Account Name', 'Category'])['Amount'].sum().sort_values(ascending=False))

print("Amount spent by account and category:")
print(by_account_category)
```

Each account plays a specific role in the budget:

**Checking Account** — used exclusively for major fixed payments: mortgage ($24,754), utilities ($2,776), mobile service ($1,680), and internet ($1,570). These are fixed expenses that remain consistent month to month.

**Platinum Card** — the primary card for everyday spending: shopping ($1,973), groceries ($1,489), dining out ($1,299), and fuel ($947). It is also used for entertainment, subscriptions, and coffee.

**Silver Card** — used alongside the Platinum card, mainly for dining out ($1,313) and groceries ($1,305). It essentially duplicates the Platinum Card’s role in the food and dining category.

## Final Observations
### 1. Two main expenses dominate the entire budget
Mortgage ($24,754) and Home Improvement ($19,092) together account for about 72% of all expenses over the 21-month period. All other categories — such as groceries, dining out, and transport — pale in comparison. When planning the budget, these two items should be considered separately from daily spending.
### 2. The budget is stable, with two anomalous months
19 out of 21 months ended with a surplus of approximately $1,700–$2,000. May 2018 (-$6,507) and June 2019 (-$6,912) are sharp outliers — likely due to major Home Improvement costs, where the average transaction is $530.
### 3. Food is a hidden major expense category
Combining groceries, restaurants, coffee, and fast food results in a total of $5,931, making it the third-largest expense after the mortgage and home repairs. Since food spending is split between the Platinum and Silver cards, it is more difficult to track and control.
### 4. Each account serves a specific role

* Checking: Used strictly for mandatory fixed payments.
* Platinum Card: The primary card for daily expenses.
* Silver Card: Duplicates the Platinum Card in food and dining categories without a clear reason for the split.

### 5. Friday and Thursday are the "most expensive" days of the week
The average transaction on Friday ($183) is three times higher than on Saturday ($25). This is counterintuitive, as one might expect spending to peak over the weekend. It is likely that large scheduled payments occur at the end of the work week.
