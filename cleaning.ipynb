# 2. Cleaning
Obj: prepare dataset to EDA

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Admin\personal-finance-analysis\data\personal_transactions.csv")
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["Month"] = df["Date"].dt.to_period("M")
df["Year"] = df["Date"].dt.year
df["Weekday"] = df["Date"].dt.day_name()
```

```python
df_clean = df[df["Category"] != "Credit Card Payment"].copy() 
```

163 rows where the payment category was “Credit Card Payment” were removed, as this category would have interfered with the analysis

```python
expenses = df_clean[df_clean["Transaction Type"] == 'debit'].copy()
income = df_clean[df_clean["Transaction Type"] == 'credit'].copy()
```

```python
print(f"Initial dataset lines: {len(df)}")
print(f"Cleaned dataset lines: {len(df_clean)}")
print(f"Number of expense transactions: {len(expenses)}")
print(f"Number of income transactions: {len(income)}")
print(f"Gaps in expenses: {expenses.isnull().sum().sum()}")
```

Cleaning results:
- "Date" converted to datetime, extracted Month/Year/Weekday
- Deleted 163 lines 'Credit Card Payments'
- For analysis: 617 expense transactions, 46 income transactions
- No gaps found - the dataset is ready for EDA
